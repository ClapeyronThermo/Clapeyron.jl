```julia
T_span = [74.1, 60.7, 42.3]; T_span .+= 273.15;
```

```julia
x_span = vcat(range(0.,0.9,91),range(0.9,1.0,51));
```

```julia
bp = bubble_pressure(model, T_span[1], [0.5,0.5])
dp = dew_pressure(model, T_span[1], [0.5,0.5]);
```

```julia
bp
```

```julia
dp
```

```julia
bps = []
dps = []

for T in T_span
    bp_T = []
    dp_T = []
    for x_co2 in x_span
        x = [x_co2, 1. - x_co2]
        bp = bubble_pressure(model, T, x)
        dp = dew_pressure(model, T, x)
        push!(bp_T, bp[1])
        push!(dp_T, dp[1])
    end
    push!(bps, bp_T)
    push!(dps, dp_T)
end
```

```julia
using PyPlot; const plt = PyPlot;
```

```julia
for (i, T) in enumerate(T_span)
    plt.plot(x_span, bps[i] ./ 1e5, label="bp"*string(T_span[i]))
    plt.plot(x_span, dps[i] ./ 1e5, label="dp"*string(T_span[i]))
end
plt.legend()
```
